# Task 1 â€“ Identify Missing Values

```python
# Import necessary libraries
!pip install missingno
import pandas as pd
import missingno as msno
```

```python
# Load the dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Task: Identify missing values in the dataset

# Step 1: Check for missing values in each column
# Hint: Use the isnull() function combined with the sum() function
missing_values_count = df.isnull().sum()

# Step 2: Print the count of missing values for each column
print("Count of missing values per column:")
print(missing_values_count)
```

```python
# Use missingno's bar function to visualize the completeness of the dataset

msno.bar(df)
```

# Task 2 â€“ Handle the Missing Values

```python
# Task: Handle missing values in the dataset


# Step 1: Replace missing values in numerical columns with their mean

# Your code here for replacing missing values in numerical columns

df['math score'].fillna(df['math score'].mean(), inplace=True)
df['reading score'].fillna(df['reading score'].mean(), inplace=True)
df['writing_score'].fillna(df['writing_score'].mean(), inplace=True)


# Step 2: Replace rows with missing values in categorical columns

# Your code here for replacing missing values in categorical columns
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
df['race/ethnicity'].fillna(df['race/ethnicity'].mode()[0], inplace=True)
df['parental level of education'].fillna(df['parental level of education'].mode()[0], inplace=True)
df['lunch'].fillna(df['lunch'].mode()[0], inplace=True)
df['test preparation course'].fillna(df['test preparation course'].mode()[0], inplace=True)

# Final Step: Verify there are no missing values left

final_missing_values_count = df.isnull().sum(0)
print("Final count of missing values per column:")
print(final_missing_values_count)


```
