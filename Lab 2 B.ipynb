# Task 4

```python
# Use the binning method in  the given dataset for the 'reading score' column.

# Apply the frequency based binning with 4 quantiles
df['NewReadingScore'] = pd.qcut(df['reading score'], q = 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Display the new discretized column
print("before binning", df['reading score'].head(10).to_string())
print("After binning", df['NewReadingScore'].head(10).to_string())

# Display the dicretized column using the boxplot
sns.boxplot(x = df['NewReadingScore'])
plt.title('Box Plot of reading score column after binning')
plt.show()
```
